<b>Task 2: Loading and Preprocessing Corpus Data</b>

```python
import re 
from collections import Counter,defaultdict
```

a) Open and read the file 

```python

file =  open('sq-sample.txt',encoding="utf8")
text =file.read()
print(text[:300])
```

b) Tokenize 

```python
tokens = re.split(r'\s+', text) 
print(f'Tokens :{(tokens[:50])}')
```

c) clean the tokens

```python
tokens_list = []
for token in tokens:
        token = token.lower()
        if re.fullmatch(r'[a-zëç]+', token): 
            tokens_list.append(token)
    

print(f'Total   tokens: {len(tokens_list)}')
print( tokens_list[:50])
```

<b>Task 3: Populating Data Structures</b>

a) Calculate token frequencies

```python
def count_token_frequencies(tokens):
    token_counter = Counter(tokens)
    return token_counter

token_frequencies = count_token_frequencies(tokens_list)
print(token_frequencies.most_common(50))
```

b) Split in to chunks (Calculate the size of the chunk and Remove overhanging tokens)

```python
num_chunks=10
chunk_size = len(tokens) // num_chunks  
chunks = [tokens[i * chunk_size: (i + 1) * chunk_size] for i in range(num_chunks)]
   
```

Calculate the token frequency for each chunk

```python
def count_frequencies_per_chunk(chunks):
    frequency_counters = []
    for chunk in chunks:
        counter = count_token_frequencies(chunk)
        frequency_counters.append(counter)
    return frequency_counters
   
```

```python
frequency_counters = count_frequencies_per_chunk(chunks)

for i, counter in enumerate(frequency_counters):
    print(f'Frequencies for Chunk {i + 1}: {counter[0]}')
```

C)  Use defaultdict

```python

def map_tokens_to_chunk_ids(frequency_counters, top_n=1250):
    token_map = defaultdict(set)
    
    for chunk_id, counter in enumerate(frequency_counters):
        top_tokens = counter.most_common(top_n)
        
        for token, _ in top_tokens:
            token_map[token].add(chunk_id)  
            
    return token_map

token_to_chunk_map = map_tokens_to_chunk_ids(frequency_counters)

for token, chunk_ids in token_to_chunk_map.items():
    print(f'Token: {token}, Chunk IDs: {chunk_ids}')
```

<b> Task 4: Answering Questions about the Data, Stumbling into a Problem</b>

a)Number of Tokens in the Top-1250 Across All Chunks

```python
top_n=1250
top_tokens_list = []
unique_top_1250_tokens = set()
for counter in frequency_counters:
    top_tokens = {token for token, _ in counter.most_common(top_n)}
    top_tokens_list.append(top_tokens)
    unique_top_1250_tokens.update(top_tokens)

total_unique_top_1250_tokens = len(unique_top_1250_tokens)
print(total_unique_top_1250_tokens)
```

b) Tokens Unique to Each Chunk’s Top-1250 List

```python

unique_top_tokens_per_chunk = []
top_1250_per_chunk = [set(counter.most_common(1250)) for counter in frequency_counters]

for i, top_tokens in enumerate(top_1250_per_chunk):
    other_chunks_top_tokens = set().union(*[top_1250_per_chunk[j] for j in range(len(top_1250_per_chunk)) if j != i])
    
    unique_tokens = top_tokens - other_chunks_top_tokens
    unique_top_tokens_per_chunk.append(unique_tokens)

for i, unique_tokens in enumerate(unique_top_tokens_per_chunk):
    print(f"Chunk {i+1} unique top-1250 tokens: {len(unique_tokens)} tokens")
    # print(unique_tokens)    
```
